**ข้อตกลง**
ถึงแม้จะเป็นงานกลุ่ม แต่ตามแนวคิดของการจัดระบบการเรียนรู้ในประเทศที่พัฒนาแล้วยอมรับ เราจะต้องให้แต่ละคนลงมือคิดลงมือทำด้วยตัวเองก่อน จากนั้นจึงค่อยมาตรวจคำตอบจากแต่ละคนในกลุ่ม หากในกลุ่มมีใครที่เร็วกว่าเพื่อนเป็นพิเศษ คนนั้นจะต้องเปิดโอกาสให้เพื่อนได้ลองค้นหาคำตอบด้วยตัวเองก่อน จะไม่บอกเพื่อนร่วมกลุ่มจนกว่าเขาจะได้พยายามแล้วและเพื่อนเป็นคนถามหาคำตอบหรือคำใบ้เอง

ที่ต้องทำอย่างนี้เพื่อให้แต่ละคนได้มีโอกาสพัฒนาทักษะในการอ่าน การวิเคราะห์ และได้ลองผิดลองถูกด้วยตนเอง ซึ่งขั้นตอนเหล่านี้แท้จริงก็มีความสนุกเพลิดเพลินอยู่ในตัวของมันเองระดับหนึ่ง เพราะแบบฝึกหัดนี้ไม่ใช่การทดสอบว่าแต่ละคนเข้าใจเนื้อหาหรือไม่ แต่เป็นการให้แต่ละคนได้ค้นพบความรู้ที่ซ่อนอยู่ในเนื้อหารายวิชาต่างหาก 

ในตอนแรกของการเดินทางจะรู้สึกหวาดกลัว ไม่อยากทำ แต่เมื่อเริ่มเข้าใจกระบวนการค้นพบความรู้แล้ว เราถึงจะเข้าใจอรรถรสในการศึกษาได้ การรีบบอกคำตอบจะไม่ต่างอะไรกับสปอยล์หนัง ดังนั้นในงานกลุ่มนี้ ขอให้มีข้อตกลงกันว่า
1. ทุกคนเริ่มทำแบบฝึกหัดด้วยตนเอง จะแยกทำที่บ้านหรือเรียกมาสุมหัวทำพร้อมกันก็ได้
2. ให้หยุดเช็กสภาพความก้าวหน้าทุกข้อย่อย คือเมื่อทำข้อ 1 เสร็จแล้วให้ตรวจคำตอบกับเพื่อนในกลุ่มด้วยว่าเข้าใจข้อ 1 ตรงกันหรือไม่ แล้วทำข้อ 2 ข้อ 3 ไปในรูปแบบเดียวกัน ถ้ามีคนช้า ให้รอคนช้าได้คิดด้วยตนเองก่อนสักพัก อย่าเร่งอีกฝ่ายมากเกินไป และคนช้าอย่ามัวแต่ social ให้ใส่ใจกับงานให้เต็มที่ ไม่งั้นทีมจะแตก
3. เนื้อหาตรงไหนถ้าทำได้แต่ไม่เข้าใจ (คือโค้ดมันออก แต่ไม่รู้ว่าทำไม) เราควรจะหาคำอธิบายตัวเองให้ได้ก่อน ถ้าไม่ได้จึงค่อยปรึกษาเพื่อนเพื่อหาคำอธิบาย
4. ถ้าในกลุ่มติดแบบยกทีม ไม่มีใครทำได้ ให้ปรึกษากันในกลุ่มก่อน ว่าแต่ละคนรู้และทดลองอะไรมาบ้าง เข้าใจแต่ละส่วนว่าอย่างไร ให้อธิบายความเข้าใจของตัวเองให้เพื่อนในกลุ่มรู้ก่อน จะได้แก้ไขความเข้าใจผิดของสมาชิกในกลุ่มให้เรียบร้อย 
5. อย่าบอกคำตอบเพื่อนจนกว่าจะถึงเวลาอันสมควร และก่อนจะบอก ขอให้คนที่อยากรู้คำตอบพยายามอธิบายว่าตัวเองรู้และเข้าใจอะไรบ้างให้เพื่อนฟัง และคนที่รู้คำตอบไม่จำเป็นต้องบอกเพื่อนตรง ๆ แต่อาจจะถามคำถามเพื่อให้เพื่อนคิดไปในทิศทางที่มีประโยชน์มากขึ้น เพราะบางทีคุณอาจจะพบว่าเพื่อนเข้าใจผิดเพียงเล็กน้อย ถ้าผ่านจุดนั้นไปได้ ที่เหลือเขาจัดการได้เอง
6. การปรึกษาข้ามกลุ่มสามารถทำได้ แต่ไม่ใช่ว่าจะเป็นการลอกกัน และต้องปรึกษาน้อย ๆ คิดและค้นหาคำตอบด้วยตัวเองมาก ๆ

**ขยายความข้อตกลงที่ 2** คนที่ทำเสร็จเร็ว ไม่จำเป็นต้องรอเพื่อก่อนลุยข้อต่อ ๆ ไป แต่ต้องเปิดโอกาสให้ทีมมาเช็คคำตอบหรือหาคำใบ้คำอธิบายต่าง ๆ เป็นระยะเพื่อปรับระดับความเข้าใจให้เพียงพอต่อการทำข้อต่อไปได้ด้วยตัวเองทุกคน 

ซึ่งขั้นตอนการพักเพื่อปรับระดับความเข้าใจนี้ ตามศัพท์การเรียนรู้เราเรียกว่า 'Leveling' เป็นกลไกที่จะทำให้ทุกคนในทีมเดินหน้าไปสู้กับปัญหาในด่านต่อไปด้วยกันได้ ถ้าไม่ปรับระดับจะมีคนถูกทิ้งไว้ข้างหลังและกลายเป็นชาวเกาะไป และเพื่อนจะรู้สึกว่าตัวเองด้อยค่าไปเรื่อย ๆ ทั้งที่ก็มีศักยภาพในการแก้ปัญหาติดตัวอยู่

### รู้จักกับคำว่า dimension และจำนวน dimension
คำว่า dimension เป็นศัพท์เทคนิคที่สำคัญมากสำหรับการใช้งาน ndarray ให้ได้คล่องแคล่วและถูกต้อง ซึ่งหากเราเปิดพจนานุกรมดู เราจะพบว่า dimension แปลว่า 'มิติ' เหมือนกับคำว่า 3D ที่ที่เราแปลว่า 3 มิติ ตัว D ในที่นี้ก็คือคำว่า Dimension 

ส่วนคำว่า dimension ของอาเรย์และงานออกแบบสถาปัตยกรม คำว่ามิตินี้แปลว่าขนาด ซึ่งถ้าหากเราเห็นตัวเลขจาก shape ว่าเป็น (50000, 32, 32, 3) ตัวเลขในทูเปิลนี้ก็คือขนาดของอาเรย์นั่นเอง โดยมิติที่ 1 มีขนาดเป็น 50,000 และมิติ 2, 3, และ 4 มีขนาดเป็น 32, 32 และ 3 ตามลำดับ

ส่วนคำว่าจำนวน dimension ก็คือจำนวนเลขระบุขนาดหรือจำนวนมิติของอาเรย์ ซึ่งตามตัวอย่างข้างบนก็คือ 4 มิติ และเราจะเรียกอาเรย์นั้นว่าอาเรย์ 4 มิติ

เรื่องน่าสนใจก็คือว่า ndarray นั้นจะเป็นอาเรย์กี่มิติก็ได้ และตัว n ใน ndarray นี้ก็คือคำย่อของ n dimensional array ซึ่งแปลว่า 'อาเรย์ n มิติ' นั่นเอง
