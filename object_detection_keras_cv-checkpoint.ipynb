## Object detection introduction

การทำ Object Detection นั้นมีกระบวนการ ระบุประเภท, จำแนกประเภท, และระบุตำแหน่งของวัตถุภายในภาพที่กำหนด โดยทั่วไปแล้ว ข้อมูลนำเข้าของคุณคือ image และ label คือกล่องที่มีป้ายกำกับ class มาให้ด้วย

การตรวจจับวัตถุสามารถคิดได้ว่าเป็นการนำ classification มาต่อยอด แต่แทนที่จะเป็น 1 label ต่อ 1 image คุณต้องตรวจจับและระบุตำแหน่งของ class ต่างๆ ได้ตามต้องการ

**ตัวอย่าง:**

<img width="250" src="https://i.imgur.com/8xSEbQD.png">

data ที่มีจะมีหน้าตาเป็นเช่นนี้ ที่เป็น รูปภาพ พร้อมกับ กล่องที่บ่งบอกตำแหน่งและ class ของวัตถุ:

```python
image = [height, width, 3]
bounding_boxes = {
  "classes": [0], # 0 is an arbitrary class ID representing "cat"
  "boxes": [[0.25, 0.4, .15, .1]]
}
```

เลขใน bounding box บอกว่า กล่องแรกนั้นคือ class0 (cat) ที่อยู่ตำแหน่ง x: 25%, y:40%, width: 15%, และ height: 10% ของ image
การเขียน format แบบนี้เรียกว่า `"rel_xywh"` "" (relative xywh) ที่ตำแหน่งสัมพันธ์กับขนาดของภาพ อย่างไรก็ตามมี format แบบอื่นด้วยเช่น `"xywh"`"" ที่เป็นค่าตัวเลขตรงๆเลยว่า xy อยู่ที่ตำแหน่งเท่าไหร่ และ w h กว้างยาวเท่าไหร่ (มักพบบ่อยกว่า)

**เรียก YOLO**

เนื่องจาก  [*You Only Look Once*](https://arxiv.org/abs/1506.02640)(YOLO) นั้นเป็นโมเดลที่มีชื่อเสียงมากในช่วงนี้ เราจะไปยืมโมเดล pre-trained ของเขามาใช้งานในตัวอย่างนี้เลย
Keras ได้เตรียมโมเดล object detection เหล่านี้ไว้ให้แล้วใน `keras_cv.models` และเราสามารถเรียกใช้งาน YOLO ได้ผ่าน `keras_cv.models.YOLOV8Detector`ซึ่งเป็นโมเดลที่เขา train มาให้แล้วนั่นเอง

** YOLO ในตัวอย่างนี้เป็นโมเดลที่ทำการ train กับ dataset ที่เรียกว่า `pascalvoc`**

````python
pretrained_model = keras_cv.models.YOLOV8Detector.from_preset(
    "yolo_v8_m_pascalvoc", bounding_box_format="xywh"
)
````

**คำอธิบาย format และ ค่า bounding box**

เห็น argument `bounding_box_format` หรือไม่

Format `xywh` นั้นเป็นการบอกโมเดลว่าค่าใน box นั้นคือค่าตำแหน่งจริงๆ (exact location) <br>
ยกตัวอย่างเช่น, ถ้ากล่องนั้นอยู่ที่ตำแหน่ง xy: (100, 100) และมีความกว้าง 55 และสูง 70 box นั้นจะมีค่าเป็น

```
[100, 100, 55, 75]
```

แต่ถ้าเป็น format แบบ `xyxy` ค่า box จะเป็น:

```
[100, 100, 155, 175]
```

สามารถเข้าไปดู `bounding_box_format` ทั้งหมดที่ keras มีให้ได้ที่ [docs](https://keras.io/api/keras_cv/bounding_box/formats/).
การตั้ง format ไม่มีผิดไม่มีถูก สามารถเลือกใช้ได้ตามต้องการ แต่ต้องเข้าใจว่า dataset ที่ดึงมานั้นใช้ format แบบใด

- ใช้ดึงภาพจาก URL ใดๆ ณ ตอนนี้ภาพที่ดึงมาเป็นภาพของ keras
- นำภาพที่ดึงออกมานั้นเก็บใส่ตัวแปรชนิด numpy array
- นำภาพนั้นมาแสดงผล โดยใช้ฟังก์ชัน `visualization` ของ keras_cv

***การปรับเกณฑ์ NonMaxSuppression***

เราสามารถปรับได้ผ่านฟังก์ชัน `keras_cv.layers.NonMaxSuppression()` 
โดยเราจะปรับให้เกณฑ์ `iou_threshold = 1.0` และ `confidence_threshold = 0.0` การปรับแบบนี้หมายความว่า "เอากล่องที่ครอบวัตถุครบและเอาทุกกล่อง"

เมื่อทำการปรับเป็นที่เรียบร้อยแล้วนำ NonMaxSuppression ที่เราแก้ไข้แล้วนั้น ไปใส่ใน `prediction_decoder` argument ในจังหวะที่ดึงโมเดลมาใส่ตัวแปร <br>
จริงๆแล้วโมเดล YOLOV8Detector นั้นมี `prediction_decoder` อยู่แล้วเราเพียงแต่เอาอันใหม่ใสว่เข้าไปแทนที่เพื่อปรับการแสดงผล

จากนั้นก็ทำการ predict แล้ว visualise เหมือนเดิม

````python
# The following NonMaxSuppression layer is equivalent to disabling the operation
prediction_decoder = keras_cv.layers.NonMaxSuppression(
    bounding_box_format="xywh",
    from_logits=True,
    iou_threshold=1.0,
    confidence_threshold=0.0,
)
pretrained_model = keras_cv.models.YOLOV8Detector.from_preset(
    "yolo_v8_m_pascalvoc",
    bounding_box_format="xywh",
    prediction_decoder=prediction_decoder,
)

y_pred = pretrained_model.predict(image_batch)
visualization.plot_bounding_box_gallery(
    image_batch,
    value_range=(0, 255),
    rows=1,
    cols=1,
    y_pred=y_pred,
    scale=5,
    font_scale=0.7,
    bounding_box_format="xywh",
    class_mapping=class_mapping,
)
print()
````

จะเห็นได้ว่าเมื่อปรับค่า threshold ข้างต้นแล้วกล่องที่ออกมานั้นมีจำนวนเยอะขึ้นกว่าตอนแรก เหตุเกิดหลักๆเพราะเราปรับ `confidence_threshold=0` แต่ต่อให้เราจะเลือกทุกๆกล่องเลย กล่องเหล่านั้นก็ถูกกรองออกด้วย `iou_threshold = 1.0` อีกที(กล่องต้องครอบวัตถุครบเท่านั้น)
